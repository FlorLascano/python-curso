#### Datos del Cliente
Creación de variable 'Diccionario' con los datos de un cliente. Definición de funciones para tratar con el cliente.

```python
client = {                                                             #Diccionario
    'ID': 40000,
    'Nombre': 'Agustin',
    'Apellido': 'Viera',
    'DNI' : 39421790,
    'Mail': 'agus_viera@hotmail.com'}

print(client)
client_full_name = client['Nombre'] +' '+ client['Apellido']
```

```python
def saludo():                                                         #Definir una función (queda grabada).
    print('Bienvenido/a ' + client.get('Nombre') + '!')              #Con dicc.get obtenemos el valor de la clave expresada.
```

```python
saludo()                                                               #Llamar a una función
```

```python
def saldo_total(saldo_a_pagar, pagos_realizados):                      #Definimos la función con sus parámetros
    diferencia = saldo_a_pagar - pagos_realizados                      #Y establecemos el ouput de la función.
    return diferencia
```

```python
saldo_a_pagar = 85000                                                  #Le damos un valor a los parámetros.
pagos_realizados = 25000
```

```python
estado_de_cuenta = saldo_total(saldo_a_pagar, pagos_realizados)    #Creamos una variable con la que podamos llamar a la función.
print('El estado de la cuenta es de $' + str(estado_de_cuenta))
```

```python
def send_mail(estado_de_cuenta):                                        #Definimos función con un condicional.
    if estado_de_cuenta > 0: 
        print('Usted tiene un saldo impago de $' + str(estado_de_cuenta))
```

```python
send_mail(estado_de_cuenta)
```

#### Librerías
Importamos y usamos la librería 'DateTime' para funcionalidades con el cliente.

```python
import datetime as dt                                                   #Importamos la librería y le damos un alias.
```

```python
help(dt)
```

```python
print(dir(dt))
```

```python
fecha_actual = dt.date.today()                              #Formato para usar las funciones de la librería:
fecha_vencimiento = dt.date(2020, 11, 1)                       #Librería.Tipo de función.Función()

print('Fecha actual : ' + str(fecha_actual))
print('Fecha de vencimiento : ' + str(fecha_vencimiento))
```

```python
dias_restantes = fecha_vencimiento - fecha_actual           #Días restantes antes que se apliquen intereses por refinanciación.
print('Días restantes al vencimiento: ' + str(dias_restantes))
```

```python
def send_mail_vcmto(estado_de_cuenta, fecha_actual, fecha_vencimiento, dias_restantes):
    if estado_de_cuenta > 0:
        print('Usted tiene un saldo impago de $' + str(estado_de_cuenta)) 
        print('al día actual de la fecha: ' + str(fecha_actual) + '.') 
        print('El resumen vence el día: ' + str(fecha_vencimiento))
        print('Días restantes al vencimiento: ' + str(dias_restantes))
    else:
        print('Al día actual de la fecha: ' + str(fecha_actual))
        print('usted no presenta deuda.')
```

```python
print('-'*48)
send_mail_vcmto(estado_de_cuenta, fecha_actual, fecha_vencimiento, dias_restantes)
print('-'*48)
```
