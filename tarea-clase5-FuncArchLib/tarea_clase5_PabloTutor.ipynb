```python
def account_balance(balance, payment): # Función para determinar el saldo de la cuenta.
    estado_de_cuenta = (balance - payment)
    return estado_de_cuenta
```

```python
balance = 15000
payment = 10000
```

```python
estado_de_cuenta = account_balance(balance, payment)
estado_de_cuenta
```

```python
def aviso(estado_de_cuenta): # Función para mandar aviso al cliente de saldos impagos.
    if estado_de_cuenta > 0:
        print("Estimado " + customer.get ("Nombre") + " usted tiene un saldo impago de " + str(estado_de_cuenta))
```

```python
import datetime as dt
```

```python
fecha_de_hoy = dt.date.today()
fecha_de_vencimiento = dt.date(2020, 12, 5)
```

```python
def resumen(estado_de_cuenta, fecha_de_hoy, fecha_de_vencimiento): # Función para informar saldo y vencimiento.
    if estado_de_cuenta > 0:
        print("Estimado cliente usted tiene un saldo impago por $ " + str(estado_de_cuenta))
        print("Al día de hoy " + str(fecha_de_hoy))
        print("Su resumen vence el " + str(fecha_de_vencimiento))
    else:
        print("A la fecha " + str(fecha_de_hoy))
        print("No presenta deuda")
           
```

```python
resumen(estado_de_cuenta, fecha_de_hoy, fecha_de_vencimiento)
```
