Al banco le interesa obtener una lista de mails para avisar a cada cliente los protocolos de distanciamiento social que aplcará cada sucursal para la atención al cliente. Para ello vamos a limpiar las columnas que no sean necesarias y conservar solamente las siguiente:

    Nombre
    Apellido
    Dni
    Mail
    Sucursal asignada


```python
clientes_sucursales_df.drop(['Unnamed: 0',"Sexo", "Fecha Nacimiento", "Color favorito", "Lugar Nacimiento", "Estado Civil"],
                                                 inplace=True, axis=1)
```

```python
clientes_sucursales_df
```

Como además el banco planea expandir sus operaciones al exterior, nos pidieron que los nombres de las columnas estén en inglés.

```python
clientes_sucursales_df.rename(columns={
    'Nombre': 'Name',
    'Apellido': 'Surname',
    'Dni': 'GovID',
    'Sucursal': 'Agency'
}, inplace=True)
```

```python
clientes_sucursales_df
```

Para facilitar el trabajo del sector de atención al cliente, nos pidieron que construyamos un mensaje para mandar por mail a los clientes y lo guardemos en una nueva columna del DataFrame usando un nombre descriptivo. El cuerpo del mensaje es el siguiente:

    "Estimado/a [nombre]: Queremos informarle que la sucursal [sucursal], está adecuando sus protocolos para satisfacer las disposiciones del COE Provincial."


```python
clientes_sucursales_df['message'] = "Estimado/a " + clientes_sucursales_df['Name'] + " " + clientes_sucursales_df['Surname'] + ": Queremos informarle que la sucursal " + clientes_sucursales_df['Agency'] + ", está adecuando sus protocolos para satisfacer las disposiciones del COE Provincial."
```
