```python
import random #Importo random para generar números al azar

clientes_azar = random.randint (0, 2) #Para generar al azar los clientes

premios_azar = random.randint(1, 10) #Para generar al azar los premios
```

```python
for premio in premios:
    if premio == premios_azar:
        print (f"¡Felicitaciones {clients [clientes_azar][1]}! Tu premio es {premios.get(premios_azar)}")
        #No puse el break acá y anduvo igual. No como en el de la consigna que sin el break no para de andar
        #Sé que no lo planteé igual pero no entiendo el porqué
```

### Lo mismo pero con ciclo while

```python
while premios_azar <= 10 or premios_azar > 0:
    print (f"¡Felicitaciones {clients [clientes_azar][1]}! Tu premio es {premios.get(premios_azar)}")
    break #Para que una vez que salga el ganador se corte
```
